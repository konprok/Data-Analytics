```python
# Przygotuj dane do użycia w STAN
stan_data = {
    'N': len(df),
    'K': len(df.columns) - 1,
    'X': df.drop('Outcome', axis=1).values,
    'y': df['Outcome'].values
}

test_data = {
    'N': len(df_test),
    'K': len(df_test.columns) - 1,
    'X': df_test.drop('Outcome', axis=1).values,
    'y': df_test['Outcome'].values
}
```

```python
# Kompilacja modelu
sm = CmdStanModel(stan_file='code.stan')

# Dopasowanie modelu
fit = sm.sample(data=stan_data, iter_sampling=2000, chains=4, iter_warmup=1000, seed=101)

# Funkcja logit
def logit(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Przewidywanie wartości dla danych testowych
beta_samples = fit.stan_variable('beta')
alpha_samples = fit.stan_variable('alpha')

# Średnie wartości parametrów
beta_mean = np.mean(beta_samples, axis=0)
alpha_mean = np.mean(alpha_samples)
```

```python
# Przewidywania
logit_preds = np.dot(test_data['X'], beta_mean) + alpha_mean
preds = logit(logit_preds) > 0.5  # Progowanie

# Prawdziwe wartości
true_labels = test_data['y']
```

```python
# Oblicz metryki oceny modelu
accuracy = accuracy_score(true_labels, preds)
precision = precision_score(true_labels, preds)
recall = recall_score(true_labels, preds)
f1 = f1_score(true_labels, preds)
conf_matrix = confusion_matrix(true_labels, preds)
```

```python
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
print("Confusion Matrix:")
print(conf_matrix)
```

```python
# Wizualizacja macierzy pomyłek jako mapa cieplna
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Diabetes', 'Diabetes'], yticklabels=['No Diabetes', 'Diabetes'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
